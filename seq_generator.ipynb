```python
import random
import re
```

```python
def generate_seq(l):
    num_seq = ("")
    while  len(str(num_seq)) != l:    
        rand = random.randint(1, 4)
        num_seq = str(num_seq) + str(rand)
    seq =re.sub(r"1", "A", str(num_seq))   
    seq =re.sub(r"2", "T", seq)    
    seq =re.sub(r"3", "G", seq)   
    seq =re.sub(r"4", "C", seq)    
    return seq
```

```python
print("seq100 = ",generate_seq(100), end="\n")
print("seq1000 = ", generate_seq(1000), end="\n")
print("seq100000 = ", generate_seq(100000), end="\n")
```
